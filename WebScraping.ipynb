```
from bs4 import BeautifulSoup
```

```
import requests
```

```
url='https://en.wikipedia.org/wiki/Saprotrophic_nutrition'
```

```
# we are sendig request to the url and taking the response
response=requests.get(url)
response
```

```
response.text
```

```
#  parsing library - html.parser and second is lxml
soup=BeautifulSoup(response.text, "lxml")
```

```
soup
```

```
for Alink in soup.find_all('a'):
  # print(Alink)
  print(Alink.get('href'))
```

```
for ptag in soup.find_all('p'):
  # print(ptag)
  # now second loop for b tag
  
  for btag in ptag.find_all('b'):
    print(btag.text)
```
